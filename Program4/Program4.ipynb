```python
import requests
from bs4 import BeautifulSoup
```

```python
links = ['https://en.wikipedia.org/wiki/History_of_Islam', 'https://en.wikipedia.org/wiki/Muhammad', 'https://en.wikipedia.org/wiki/2024_ICC_Men%27s_T20_World_Cup', 'https://en.wikipedia.org/wiki/2026_FIFA_World_Cup']
documents = []
unique_words = set()

for link in links:
    r = requests.get(link)
    soup = BeautifulSoup(r.content, 'html.parser')
    s = soup.find('div', class_='mw-body-content')
    paras = s.find_all('p')
    paragraphs = ""
    for line in paras:
        text = re.sub(r'\[.*?\]', '', line.text)
        text = re.sub(r'[^\x00-\x7F]+', '', text)
        text = re.sub(r'[\\]+','', text)
        text = re.sub('\n','',text)
        text = re.sub(r'[,."\'/(){}:;]', '', text)
        if len(text) != 0:
            unique_words.update(word_tokenize(text.lower()))
            paragraphs += text.lower()
    documents.append(paragraphs)

len(unique_words)
```

```python
for i, doc in enumerate(documents):
    open(f'sample_{i}.txt', 'w').write(doc)
```

Unigram matrix

```python
tokens = [document.split() for document in documents]

vocabulary = list(set(word for document in tokens for word in document))

uni_count_matrix = [[0] * len(vocabulary) for _ in range(len(documents))]

for i, document in enumerate(tokens):
    for word in document:
        if word in vocabulary:
            uni_count_matrix[i][vocabulary.index(word)] += 1

print("Unigram Count Matrix:")
for row in uni_count_matrix:
    print(row)

print("\nVocabulary:")
print(vocabulary)
```

Bigram matrix

```python
bigrams = [document.split() for document in documents]
bigrams = [[' '.join(bigrams[i][j:j+2]) for j in range(len(bigrams[i])-1)] for i in range(len(bigrams))]
# Create a flat list of bigrams
flat_bigrams = [bigram for document_bigrams in bigrams for bigram in document_bigrams]
# Create a vocabulary of unique bigrams
vocabulary = list(set(flat_bigrams))
# Create an empty count matrix
bi_count_matrix = [[0] * len(vocabulary) for _ in range(len(documents))]
# Populate the count matrix
for i, document_bigrams in enumerate(bigrams):
    for bigram in document_bigrams:
        if bigram in vocabulary:
            bi_count_matrix[i][vocabulary.index(bigram)] += 1

print('Bigram count matrix')
for row in bi_count_matrix:
    print(row)

print('Vocabulary')
print(vocabulary)
```

TF IDF count matrix
TF(t) = (Number of times term t appears in a document) / (Total number of terms in the document)
IDF(t) = (Total number of documents in a collection) / (Number of documents in the collection that contain the term t)

```python
def calculate_term_frequency(document):
    term_frequency = {}
    total_terms = len(document)

    for term in document:
        term_frequency[term] = term_frequency.get(term, 0) + 1

    for term, count in term_frequency.items():
        term_frequency[term] = count / total_terms

    return term_frequency

def calculate_inverse_document_frequency(corpus, term):
    document_count = sum(1 for document in corpus if term in document)
    if document_count > 0:
        return math.log(len(corpus) / document_count)
    else:
        return 0.0
```

```python
tokens = [document.split() for document in documents]

# Get unique terms (words) from the corpus
vocabulary = list(set(word for document in tokens for word in document))

# Initialize TF-IDF matrix
tfidf_matrix = []

# Calculate TF-IDF values for each document and term
for document in tokens:
    tfidf_vector = []
    term_frequency = calculate_term_frequency(document)

    for term in vocabulary:
        tf = term_frequency.get(term, 0)
        idf = calculate_inverse_document_frequency(tokens, term)
        tfidf_vector.append(tf * idf)

    tfidf_matrix.append(tfidf_vector)

# Display the TF-IDF matrix and vocabulary
print("TF-IDF Matrix:")
for row in tfidf_matrix:
    print(row)

print("\nVocabulary:")
print(vocabulary)
```

Naive Bayes

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
labels = [0,0,1,1]

# Split the data into training and testing sets
matrices = [uni_count_matrix, bi_count_matrix, tfidf_matrix]
matrix_labels = ['Unigram', 'Bigram', 'Tf-IDF']

for i in range(3):
    X_train, X_test, y_train, y_test = train_test_split(matrices[i], labels, test_size=0.25, random_state=42)

    classifier = MultinomialNB()
    classifier.fit(X_train, y_train)

    predictions = classifier.predict(X_test)

    accuracy = metrics.accuracy_score(y_test, predictions)
    precision = metrics.precision_score(y_test, predictions)
    recall = metrics.recall_score(y_test, predictions)

    print(f"\n\nPerformance in {matrix_labels[i]} matrix")
    print("Accuracy:", accuracy)
    print("Precision:", precision)
    print("Recall:", recall)
```
